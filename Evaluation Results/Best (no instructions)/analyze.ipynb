```python
import json
#load the predicted json
predicted_json = dict()
with open('predicted_specs_tests_minprompt.json', 'r' , encoding='utf-8') as f:
    predicted_json = json.load(f)
```

```python
# correctly format predicted json's predicted_text column to remove "\"
for item in predicted_json:
    item['predicted_text'] = item['predicted_text'].replace('\\', '')

# 2. Save back to a file
with open('cleaned_predictions.json', 'w', encoding='utf-8') as f:
    json.dump(predicted_json, f, indent=4, ensure_ascii=False)
```
