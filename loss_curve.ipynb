```python
import matplotlib.pyplot as plt
import numpy as np

# Chart types
chart_types = [
    "Area", "Bar", "Box", "Histogram", "Line", "Pie", "Scatter"
]

# ===== DATA =====
baseline_topology = [92.5, 46.0, 84.6, 88.9, 45.3, 100.0, 79.2]
finetuned_strict_topology = [95.0, 56.0, 84.6, 88.9, 70.7, 100.0, 94.3]
finetuned_minimal_topology = [95.0, 86.0, 86.7, 100.0, 98.4, 100.0, 90.3]

baseline_domain = [69.8, 65.9, 74.1, 47.2, 63.6, None, 75.9]
finetuned_strict_domain = [87.9, 69.6, 58.9, 80.6, 80.9, None, 77.1]
finetuned_minimal_domain = [96.5, 92.6, 83.7, 97.2, 84.5, None, 85.7]

baseline_categorical = [45.2, 70.9, None, 18.4, 73.6, 97.5, 71.4]
finetuned_strict_categorical = [67.2, 70.1, 56.1, 42.8, 75.1, 97.6, 76.1]
finetuned_minimal_categorical = [77.0, 88.2, 74.2, 60.0, 86.9, 96.4, 80.6]

baseline_semantic = [15.5, 24.3, 8.7, 0.0, 58.7, None, None]
finetuned_strict_semantic = [18.0, 51.0, 36.6, 0.0, 58.4, None, None]
finetuned_minimal_semantic = [66.9, 77.6, 50.9, 81.1, 82.6, None, 63.0]


def plot_metric(metric_name, base, strict, minimal):
    x = np.arange(len(chart_types))
    width = 0.25

    base = np.array([np.nan if v is None else v for v in base])
    strict = np.array([np.nan if v is None else v for v in strict])
    minimal = np.array([np.nan if v is None else v for v in minimal])

    plt.figure()

    plt.bar(x - width, base, width, label="Baseline (Strict)")
    plt.bar(x, strict, width, label="Finetuned (Strict)")
    plt.bar(x + width, minimal, width, label="Finetuned (Minimal Prompt)")

    plt.xticks(x, chart_types)
    plt.ylabel("Score (%)")
    plt.title(f"{metric_name} Comparison by Chart Type")
    plt.legend()
    plt.grid(axis="y")

    plt.tight_layout()
    plt.show()


# ===== PLOTS =====

plot_metric("Topology Accuracy",
            baseline_topology,
            finetuned_strict_topology,
            finetuned_minimal_topology)

plot_metric("Domain Accuracy",
            baseline_domain,
            finetuned_strict_domain,
            finetuned_minimal_domain)

plot_metric("Categorical Accuracy",
            baseline_categorical,
            finetuned_strict_categorical,
            finetuned_minimal_categorical)

plot_metric("Semantic Score",
            baseline_semantic,
            finetuned_strict_semantic,
            finetuned_minimal_semantic)
```
